```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
# Create new_df and 'tags' column from the 'movies' DataFrame
new_df = movies[['id', 'original_title', 'genres', 'overview', 'cast', 'director', 'keywords']].copy()

# Process string columns for tag creation
new_df['genres'] = new_df['genres'].str.replace('|', ' ')
new_df['keywords'] = new_df['keywords'].str.replace('|', ' ')
new_df['cast'] = new_df['cast'].str.replace('|', ' ')
new_df['director'] = new_df['director'].str.replace('|', ' ')# Director is a single name, just lowercase
new_df['overview'] = new_df['overview'].str.replace('|', ' ')

# Combine relevant columns into a single 'tags' column
new_df['tags'] = new_df['overview'] + " " + new_df['genres'] + " " + new_df['keywords'] + " " + new_df['cast'] + " " + new_df['director']

# Original line that caused the error, now new_df and 'tags' are defined
vectors = cv.fit_transform(new_df['tags']).toarray()
```
